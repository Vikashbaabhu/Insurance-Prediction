#4. Divide the data into input and output

```python
# Input - age, sex, bmi, children, Employment, vehicles, Years, Interest, Rate
# Output - charges
```

```python
# Input
x = df.iloc[:,0:8].values 
x
```

```python
# Output 
y = df.iloc[:,8].values
y
```

#5. Train and Test variables

```python

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 0)
```

```python
#After train_test_split ,

# Training varaibles get 75% data of x and y
# Testing variables get 25% data from x and y
```

```python
print(x.shape) 
print(x_train.shape)
print(x_test.shape)
```

```python
print(y.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# We don't perform step 6 here. 
```

#7. Run Linear Regression 

```python

from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

#8. Fit the model

```python

model.fit(x_train,y_train)
```

#9. Predict the output

```python

y_pred = model.predict(x_test) 
y_pred 
```
